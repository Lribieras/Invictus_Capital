```python
import numpy as np
import pandas as pd
import math
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import axes
import pickle
from datetime import datetime
from scipy import log,exp,sqrt,stats
print('Packages')
```

```python
returns_2x = pd.read_csv('/Users/louisribieras/Desktop/2x_varying_caps.csv')
returns_3x = pd.read_csv('/Users/louisribieras/Desktop/3x_varying_caps.csv')
c10 = pd.read_csv('/Users/louisribieras/Desktop/c10_returns.csv')
```

```python
returns_2x = returns_2x.set_index(returns_2x['date'])
returns_3x = returns_3x.set_index(returns_3x['date'])
c10 = c10.set_index(c10['date'])
returns_2x.head(5)
```

```python
plt.ylabel('Percentage Change /100%')
plt.rcParams['figure.figsize'] = (30,15)
plt.plot(c10['c10'], label="c10")
plt.plot(returns_2x['returns 2x (25)'], label="25% cap")
plt.plot(returns_2x['returns 2x (50)'], label="50% cap")
plt.plot(returns_2x['returns 2x (75)'], label="75% cap")
plt.plot(returns_2x['returns 2x (100)'], label="100% cap")
plt.xticks(np.arange(0,365,step=7), rotation=90)
plt.title('C10 vs C10 w/Tether 2x (Varying Caps)')
plt.legend()
plt.show()
```
